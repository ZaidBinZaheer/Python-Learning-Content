```python
import os
import pandas as pd
```

```python
# load planes_features data
planes = pd.read_csv('Dataset_FlightRadar24_1.csv')
planes
```

```python
planes['Time'] = planes['UTC'].str[11:19]
planes
```

##### Grouping data based on Flight_ID

```python
grouped = planes.groupby('Flight_ID')
```

```python
import matplotlib.pyplot as plt
# import seaborn as sns
```

```python

# Plot using matplotlib
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot altitude on the first y-axis (blue)
ax1.plot(planes['Time'], planes['Altitude'], color='blue', marker='o', linestyle='-', label='Altitude')
ax1.set_xlabel('Time')
ax1.set_ylabel('Altitude', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for speed (green)
ax2 = ax1.twinx()
ax2.plot(planes['Time'], planes['Speed'], color='green', marker='s', linestyle='-', label='Speed')
ax2.set_ylabel('Speed', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Add legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Altitude and Speed vs. Time')
plt.show()
```

```python
# Group the dataset by 'Flight_ID'
# grouped = planes.groupby('Flight_ID')

# Iterate over each group
for flight_id, group_data in grouped:
    print(f"Flight ID: {flight_id}")
    # print(group_data.head())  



    # Plot using matplotlib
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # Plot altitude on the first y-axis (blue)
    ax1.plot(group_data['Time'], group_data['Altitude'], color='blue', marker='o', linestyle='-', label='Altitude')
    ax1.set_xlabel('Time')
    ax1.set_ylabel('Altitude', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Create a second y-axis for speed (green)
    ax2 = ax1.twinx()
    ax2.plot(group_data['Time'], group_data['Speed'], color='green', marker='s', linestyle='-', label='Speed')
    ax2.set_ylabel('Speed', color='green')
    ax2.tick_params(axis='y', labelcolor='green')

    # Add legend
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

    plt.title('Altitude and Speed vs. Time')
    plt.show()
```
